```python
## Changing the data type from sring/objec to Date Time


df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df.info()

print('/////////////////////////////////////////////')

type(df['timeStamp'].iloc[0])
```

```python
# creatinga new 3 columns and spiting the values of Datetime-> Data type.
#inserting into the 3cloumns as Hour,Month,Day of week

T = df['timeStamp'].iloc[0]
T.hour
T.year
```

```python
# new Column for Hours

df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)

#new column for Month
df['Month'] = df['timeStamp'].apply(lambda month: month.month)

## new column for day of week in 'English'

df['Day of week'] = df['timeStamp'].apply(lambda day:day.day_name(locale ='English'))
```

```python
df.head(2)
```

```python
# new column for day number

day_num = {'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'friday':4,'Saturday':5,'Sunday':6}

df['Day num'] = df['Day of week'].map(day_num)


df.head(3)
```
