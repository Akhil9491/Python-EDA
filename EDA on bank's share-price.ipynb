```python
import yfinance as yf
import datetime
```

```python
# Define the start and end dates
start = '2006-01-01'
end = '2016-01-01'
```

```python
# Fetch the data using yfinance

# Bank of America Stock data

BAC = yf.download('BAC', start=start, end=end)
BAC.head(5)
```

```python
# now same downloading the data sets fro various investment banks

# Citi Groups

C = yf.download("C",Start, End)

# Goldman Sachs
GS = yf.download('GS', Start, End)

# Morgan Stanley

MS = yf.download('MS', Start, End)

# Wells Fargo

WF = yf.download('WF', Start, End)
```

```python
# Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers** for multi level indexing

tickers = ['BAC','C','GS','MS','WF']
```

```python
# Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks.

bank_stocks = pd.concat([BAC,C,GS,MS,WF], axis =1, keys = tickers)
```

```python
bank_stocks.head()
```

```python
# setting the column name levels

bank_stocks.columns.names = ['Bank Tickers', 'Stock Info']
```

```python
bank_stocks.head()
```

```python
# EDA

Let's explore the data a bit! Before continuing, I encourage you to check out the documentation on 
[Multi-Level Indexing]--(http://pandas.pydata.org/pandas-docs/stable/advanced.html)
and [Using .xs](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html).
Reference the solutions if you can not figure out how to use .xs(), since that will be a major part of this project.

** What is the max Close price for each bank's stock throughout the time period?**
```

```python
# maximum value of close share price in multi index 
bank_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').max()
```

```python
## creating a new DataFrame called returns


returns = pd.DataFrame()
```

```python
# Pandas dataframe.pct_change() function calculates the percentage change between the current and a prior element.
# This function by default calculates the percentage change from the immediately previous row.





for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
# seaborn pairplot for return dataframe


sns.pairplot(data = returns[1:],markers = '*')
```

```python
## Pandas dataframe.idxmax()/.idxmin() function returns index of first occurrence of maximum/minimum over requested axis. 
##While finding the index of the maximum value across any index, all NA/null values are excluded.


returns.idxmin()
```

```python
returns.idxmax()
```
